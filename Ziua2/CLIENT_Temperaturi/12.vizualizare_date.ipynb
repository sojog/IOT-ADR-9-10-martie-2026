```python
read_df = pd.read_csv("temp_incrementale.csv")
```

```python
read_df
```

```python
new_df = read_df[:10]
plt.scatter(new_df.index, new_df["temp"])
plt.plot(new_df.index, new_df["temp"])
```

```python
Q1 = new_df["temp"].quantile(0.25)
Q3 = new_df["temp"].quantile(0.75)
```

```python
DIFF = Q3 - Q1
DIFF
```

```python
lower_bound = Q1 - 1.5 * DIFF
upper_bound = Q3 + 1.5 * DIFF 
```

```python
lower_anomalies = new_df [ new_df["temp"] < lower_bound ]
lower_anomalies
```

```python
upper_anomalies = new_df [ new_df["temp"] > upper_bound ]
upper_anomalies
```
