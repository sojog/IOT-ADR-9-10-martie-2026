```python
import requests
```

```python
URL = "http://localhost:11434/api/chat"
```

```python
MODEL = "qwen2.5-coder:14b"
```

```python
PROMPT = "Explica-mi protocolul MQTT"
```

```python
first_message = {
    "role" : "user", "content" : PROMPT
}
```

```python
payload = {
    "model" : MODEL,
    "messages" : [first_message],
    "stream" : False
}
```

```python
response = requests.post(URL, json=payload)
response.content
```

```python
print(response.text)
```

```python
mesajul_primit = response.json()["message"]
mesajul_primit
```

```python
second_question =  {
    "role" : "user", "content" : "Ce-l face un protocol?"
}
```

```python
second_payload = {
    "model" : MODEL,
    "messages" : [first_message, mesajul_primit, second_question],
    "stream" : False
}
```

```python
second_payload
```

```python
response = requests.post(URL, json=second_payload)
response
```

```python
response.json()["message"]
```
