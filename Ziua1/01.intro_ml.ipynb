# Antrenarea modelului

```python
result = stats.linregress(suprafete, preturi)
result
```

```python
result.slope, result.intercept
```

```python
plt.plot(suprafete, preturi)
plt.scatter(suprafete, preturi)
plt.plot(suprafete, 2000 * suprafete  + 100, c="orange")
plt.plot(suprafete, 2500 * suprafete  + 100, c="green")

plt.plot(suprafete, result.slope * suprafete  + result.intercept, c="black")

plt.xlabel("suprafete");
plt.ylabel("Preturi")
```

# Folosirea modelului

```python
def f(x):
    return result.slope * x + result.intercept
```

```python
f(120)
```

```python

plt.scatter(suprafete, preturi)
plt.plot(suprafete, result.slope * suprafete  + result.intercept, c="black")

mp = 160
plt.scatter(mp, f(mp))

plt.xlabel("suprafete");
plt.ylabel("Preturi")
```

# SKLEARN vs SCIPY

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
suprafete
```

```python
suprafete_2d = suprafete.reshape(-1, 1)
suprafete_2d
```

```python
model.fit(suprafete_2d, preturi)
```

```python
model.coef_, model.intercept_ # rezultat SKLEARN
```

```python
result.slope, result.intercept # rezultat SCIPY
```

```python
plt.title("SCIPY")

plt.scatter(suprafete, preturi)
plt.plot(suprafete, result.slope * suprafete  + result.intercept, c="black")

mp = 160
plt.scatter(mp, f(mp))

plt.xlabel("suprafete");
plt.ylabel("Preturi")
```

```python
plt.title("SKLEARN")

plt.scatter(suprafete, preturi)
plt.plot(suprafete, model.coef_ * suprafete  + model.intercept_, c="black")

mp = 160
plt.scatter(mp, model.predict([[mp]]))

plt.xlabel("suprafete");
plt.ylabel("Preturi")
```
